`numpy.histogram`

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science','notebook'])
```

A histogram is a bar plot corresponding to a dataset $\{x_j\}$ where the height of each bar $c_i$ is given by

$$c_i = \sum_{x_j \in b_i} 1$$

where $b_i=[e_{1}, e_{2})_i$ is some interval specified by two edges $e_1$ and $e_2$.

# Coding

Obtain a bunch of random variables distributed according to a Gaussian distribution:

```python
x = np.random.randn(int(1e6))
```

```python
x[0:20]
```

Use the numpy `histogram` function to get counts and bins. The function automatically generates 10 equally spaced bins
* `c` is counts
* `e` is bin edges
* `b` is bin centers

```python
c, e = np.histogram(x)
```

To obtain the centers of the bins, we take the average of consecutive bin edges

```python
b = (e[1:]+e[:-1])/2
```

We can then plot the bin centers vs. number of counts on a scatter plot.

```python
plt.figure(figsize=(4,3))
plt.scatter(b, c)
plt.show()
```

Of course, `matplotlib` already has a function for plotting histograms given data that works exactly the same as `numpy`s histogram function. It plots a bar graph where the center of the bars are in the middle of the bins:

```python
plt.figure(figsize=(3,3))
plt.hist(x)
plt.show()
```

# Customizability

## Bin Edges

One can also specify their own bin edges:

```python
e = np.linspace(-5,5,101)
```

Use the `histogram` function, but this time specify the bin edges generated:

```python
c, e = np.histogram(x, bins=e)
b = (e[1:]+e[:-1])/2
```

Make a scatter plot of bins and bin heights

```python
plt.figure(figsize=(4,3))
plt.scatter(b, c)
plt.show()
```

## Weighted histograms

A weighted histogram becomes 

$$c_i = \sum_{x_j \in b_i} w_j$$

where $w_j$ is the "weight" of data point $x_j$. For example, suppose we had $10000$ particles where $x_j$ is the position of these particles in space:

```python
x = np.random.randn(10000)
c, e = np.histogram(x, bins=100)
b = (e[1:]+e[:-1])/2
```

The histogram below plots the number of particles in each of the bin intervals $b_i$ in space:

```python
plt.figure(figsize=(4,3))
plt.scatter(b, c)
plt.show()
```

Suppose we want to instead plot the charge distribution in space, and each particle has a charge $q_j = x_j^2$. If we wanted the total charge in space, we would then need to plot 

$$c_i = \sum_{x_j \in b_i} q_j = \sum_{x_j \in b_i} x_j^2 $$

Thus we set $w_j=x_j^2$ to get our charge distribution:

```python
x = np.random.randn(10000)
c, e = np.histogram(x, bins=100, weights=x**2)
b = (e[1:]+e[:-1])/2
```

Plot the charge distribution

```python
plt.figure(figsize=(4,3))
plt.scatter(b, c)
plt.show()
```

This is also useful in statistics, where the statistical uncertainty on different data points $x_j$ are different.

# Other Functionality

The `density` argument will normalize the histogram heights $c_i$ so that the integral under the histogram is 1

```python
x = np.random.randn(10000)
c, e = np.histogram(x, bins=100, density=True)
b = (e[1:]+e[:-1])/2
```

Compute the integral:

```python
sum(c)*np.diff(b)[0]
```

Show what this looks like:

```python
plt.figure(figsize=(4,3))
plt.scatter(b, c)
plt.show()
```

This is useful if you don't care about the absolute number of counts.

# Conclusion

Histograms are useful ways of reducing the amount of storage space required to store data. The trade off is

* You lose the order of the data points (everything becomes aggregated in bins)
* You lose precision on the data points. You only know that each data point resides between the edges of a bin

```python
x = np.random.randn(int(1e7))
```

Look at the size of this data in MB

```python
x.size*x.itemsize/ 1e6
```

In a histogram:

```python
c, e = np.histogram(x, bins=10000)
b = (e[1:]+e[:-1])/2
```

```python
b.size*b.itemsize/ 1e6
```

```python
c.size*c.itemsize/ 1e6
```

Massive reduction in amount of memory to store data, but precision is lost!
