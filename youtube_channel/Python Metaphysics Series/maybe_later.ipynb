```python
import numpy as np
from scipy.optimize import minimize, LinearConstraint, NonlinearConstraint
from scipy.integrate import quad
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

Minimize

$$\text{Min}_{a,b}\iint \left|\sqrt{x^2+y^2} - (ax+by)\right| dx dy$$


```python
x = y = np.linspace(-2,2,100)
xv, yv = np.meshgrid(x,y)
zv = np.sqrt(xv**2+yv**2)
```

```python
fig = go.Figure(data=[go.Surface(z=np.sqrt(xv**2+yv**2), x=x, y=y, colorscale='Viridis', showscale=False),
                      go.Surface(z=0.5*xv+0.5*yv, x=x, y=y, showscale=False)])
fig.update_layout(width=500, height=500,)
fig.show()
```

```python
def integrand(x, a, b):
    return (1-(a*x+b*np.sqrt(1-x**2)))**2
```

```python
def obj_ab(ab):
    return quad(integrand, 0, 1, args=(ab[0], ab[1]))[0]
```

```python
x0 = np.array([0.2, 0.5])
res = minimize(obj_ab, x0=x0, method='CG').x
```
